```python
import xml.etree.ElementTree as et
import datetime
import sys


# init:
root = et.Element("Crimes")
tree = et.ElementTree(root)
crime = et.SubElement(root, "Crime")

def end_program():
    tree.write('crime_reporter.xml', encoding='UTF-8')
    sys.exit()

next = False

while not next:
    proceed = raw_input("Welcome to the Crime Reporter database. Would like to proceed [Y/N]").upper()
    if proceed == 'Y':
        pass
    elif proceed == 'N':
        end_program()
    else
        print 'Please enter Y for YES or N for NO'
        continue
    q1 = raw_input("What type of crime are you reporting?\nPlease select [A/B/C] A) Auto Theft B) Vandelism C) Assault: ").upper()
    now = datetime.datetime.now() #.strftime("%d/%m/%y %H:%M")
#     crime.set('year', str(now.year))
#     crime.set('month', str(now.month)
#     crime.set('day', str(now.day))
#     crime.set('hour', str(now.hour))
#     crime.set('minute', str(now.minute))
    
    if q1 == 'A':
        crime.set('type', 'Auto Theft')
        next = True
    elif q1 == 'B':
        crime.set('type', 'Vandelism')
        next = True
    elif q1 == 'C':
        crime.set('type', 'Assault')
        next = True
    else:
        print "Please enter A or B or C"
        print ""
        
if crime.attrib['type'] == 'Auto Theft':
    owner_name = et.SubElement(crime, "owner_name")
    owner_name.text = raw_input("What is the name of the vehicle owner:")
    
    vehicle_make = et.SubElement(crime, 'vehicle_make')
    vehicle_make.text = raw_input("What is the vehicle make:")
    
    vehicle_model = et.SubElement(crime, 'vehicle_model')
    vehicle_model.text = raw_input("What is the vehicle model:")
    
    vehicle_color = et.SubElement(crime, 'vehicle_color')
    vehicle_color.text = raw_input("What is the color of the vehicle:")
elif crime.attrib['type'] == 'Vandelism':
    property_type = et.SubElement(crime, 'property_type')
    property_type.text = raw_input("What type of property was vandelized:")
    
    estimated_cost = et.SubElement(crime, 'estimated_cost')
    estimated_cost.text = raw_input("What is the estimated cost of damage:")
    
    witness_name = et.SubElement(crime, 'witness_name')
    witness_name.text = raw_input("Name of witness (leave blank for none):")
elif crime.attrib['type'] == Assault:
    victim_name = et.SubElement(crime, 'victim_name')
    victim_name.text = raw_input("What is the name of the victim:")
    
    assailant_name = et.SubElement(crime, 'assailant_name')
    assailant_name.text = raw_input("What is the name of the assailant:")
    
    injury = et.SubElement(crime, 'injury:')
    injury.text = raw_input("Describe the injury:")
    
    weapon = et.SubElement(crime, 'weapon:')
    weapon.text = raw_input('What type of weapon was used:')
    
et.dump(root)

```
